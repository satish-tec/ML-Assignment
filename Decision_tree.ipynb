```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

train_df = pd.read_csv('/content/drive/MyDrive/Datasets/train.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Datasets/test.csv')
```

```python

for col in ['Type', 'Price']:
    train_df[col] = train_df[col].str.strip()
    test_df[col] = test_df[col].str.strip()

train_processed = pd.get_dummies(train_df, columns=['Type', 'Price'])
test_processed = pd.get_dummies(test_df, columns=['Type', 'Price'])

train_features = train_processed.drop(['Example', 'Category'], axis=1)
test_features = test_processed.reindex(columns=train_features.columns, fill_value=0)

# Encode labels (Yes/No to 1/0)
le = LabelEncoder()
y_train = le.fit_transform(train_processed['Category'])
y_test = le.transform(test_processed['Category'])
```

```python
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(train_features, y_train)
```

```python
y_pred = clf.predict(test_features)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred):.2f}')
```
